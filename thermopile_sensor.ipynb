# DC offset removal , Spectral analysis & Histogram analysis

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
import pandas as pd
```

```python
# Load the CSV file
df = pd.read_csv('1kHz_5V.csv', index_col=0)

# Original signal properties
fs = 1000 
dc_offset = 0.35
t = df.index.values
original_signal = df['Voltage']

# Remove DC offset
signal_without_offset = original_signal - original_signal.mean()

# Remove DC offset using a high-pass filter
nyquist = 0.5 * fs
cutoff_freq = 300
order = 1

# Design a high-pass filter
b, a = signal.butter(order, cutoff_freq/nyquist, btype='high', analog=False)

# Apply the filter to remove DC offset
filtered_signal = signal.filtfilt(b, a, signal_without_offset)

# Plot the filtered signal
plt.subplot(2, 1, 2)
plt.plot(t, filtered_signal, label='1kHz Filtered Signal')
plt.title('Signal after DC Offset Removal with High-pass Filter')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()

plt.tight_layout()
plt.show()



# Spectral analysis using FFT
fft_result = np.fft.fft(filtered_signal)
fft_freq = np.fft.fftfreq(len(fft_result), d=1/fs)

# Plot the magnitude spectrum
plt.figure(figsize=(7, 3))
plt.plot(fft_freq, np.abs(fft_result))
plt.title('Spectral Analysis [fs = 1kHz]')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()



# Histogram
plt.subplot(3, 1, 3)
plt.hist(filtered_signal, bins='auto', color='blue', alpha=0.7, rwidth=0.85)
plt.title('Histogram of 1kHz Filtered Signal')
plt.xlabel('Amplitude')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
