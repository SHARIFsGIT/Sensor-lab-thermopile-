# Measurement of time constants

```python
_5V_sampleTime10s_sampleRate1kHz_sampleFrequency0Hz_withHand = pd.read_csv('5V_sampleTime10s_sampleRate1kHz_sampleFrequency0Hz_withHand.csv', index_col=0)
_5V_sampleTime10s_sampleRate1kHz_sampleFrequency0Hz_withoutHand = pd.read_csv('5V_sampleTime10s_sampleRate1kHz_sampleFrequency0Hz_withoutHand.csv', index_col=0)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize= (20,5))
fig.suptitle("Smaple time 10s with 1kHz sample frequency & shutter frequency 0Hz")

ax[0].set(title = "With blockage")
ax[0].set_xlabel("Time [s]")
ax[0].set_ylabel("Voltage [V]")
ax[0].plot(_5V_sampleTime10s_sampleRate1kHz_sampleFrequency0Hz_withHand)

ax[1].set(title = "Without blockage")
ax[1].set_xlabel("Time [s]")
ax[1].set_ylabel("Voltage [V]")
ax[1].plot(_5V_sampleTime10s_sampleRate1kHz_sampleFrequency0Hz_withoutHand)
```

```python
withBBR_alwaysOpen_withoutHand = pd.read_csv('withBBR_alwaysOpen_withoutHand.csv', index_col=0)
withBBR_alwaysOpen_withHand = pd.read_csv('withBBR_alwaysOpen_withHand.csv', index_col=0)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize= (20,5))
fig.suptitle("Always Open Shutter Mode")

ax[0].set(title = "Thermal radiation passed")
ax[0].set_xlabel("Time [s]")
ax[0].set_ylabel("Voltage [V]")
ax[0].plot(withBBR_alwaysOpen_withoutHand)

ax[1].set(title = "Thermal radiation blocked")
ax[1].set_xlabel("Time [s]")
ax[1].set_ylabel("Voltage [V]")
ax[1].plot(withBBR_alwaysOpen_withHand)
```
