```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
_1kHz_0V = pd.read_csv('1kHz_0V.csv', index_col=0)
_1kHz_5V = pd.read_csv('1kHz_5V.csv', index_col=0)
_1kHz_5V_hand = pd.read_csv('1kHz_5V_hand.csv', index_col=0)

_5kHz_0V = pd.read_csv('5kHz_0V.csv', index_col=0)
_5kHz_5V = pd.read_csv('5kHz_5V.csv', index_col=0)
_5kHz_5V_hand = pd.read_csv('5kHz_5V_hand.csv', index_col=0)

_10kHz_0V = pd.read_csv('10kHz_0V.csv', index_col=0)
_10kHz_5V = pd.read_csv('10kHz_5V.csv', index_col=0)
_10kHz_5V_hand = pd.read_csv('10kHz_5V_hand.csv', index_col=0)

_40kHz_0V = pd.read_csv('40kHz_0V.csv', index_col=0)
_40kHz_5V = pd.read_csv('40kHz_5V.csv', index_col=0)
_40kHz_5V_hand = pd.read_csv('40kHz_5V_hand.csv', index_col=0)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize= (20,5))
fig.suptitle("Smaple time 10s with 1kHz sample frequency")

ax[0].set(title = "OV")
ax[0].set_xlabel("Time [s]")
ax[0].set_ylabel("Voltage [V]")
ax[0].plot(_1kHz_0V)

ax[1].set(title = "5V")
ax[1].set_xlabel("Time [s]")
ax[1].set_ylabel("Voltage [V]")
ax[1].plot(_1kHz_5V)

ax[2].set(title = "5V with hand")
ax[2].set_xlabel("Time [s]")
ax[2].set_ylabel("Voltage [V]")
ax[2].plot(_1kHz_5V_hand)
```
